```python
import pandas as pd

data = pd.read_csv(r'C:\Users\mkrti\Downloads\ibm.csv')

data
```

```python
# Calculate descriptive statistics
descriptive_stats = data.describe()

# Extracting relevant columns for line graph
line_graph_data = data[['Date', 'Close']]

# Convert 'Date' column to datetime
line_graph_data['Date'] = pd.to_datetime(line_graph_data['Date'])

# Plotting the line graph for the 'Close' prices
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(line_graph_data['Date'], line_graph_data['Close'], marker='o')
plt.title('AAPL Stock Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.grid(True)
plt.show()

descriptive_stats
```
